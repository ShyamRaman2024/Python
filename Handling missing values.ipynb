# Handling missing values

```python
import pandas as pd 
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_excel('./windspeed.xlsx')
data.head(31)
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().mean()*100
```

### Filling a common value to all missing data

```python
data.fillna(0)
```

```python
data.fillna('NULL')
```

### Adding missing data to individual columns

```python
data.Status.fillna('sunny')
```

* We can access individual columns by an alternate wayalso.

```python
data.fillna({
    'Temperature':0,
    'WindSpeed':10,
    'Status':'sunny'
})
```

#### Forward fill (row)
* Forward fill is a method to forward the data from the row above the missing value. Thus all the missing value will get filled with the value above. If there are multiple missing valuesconsecutively , they will also get filled with the same value of the above available data.

```python
data
```

```python
data.fillna(method='ffill')
```

#### Backward fill (row)

```python
data.fillna(method='bfill')
```

#### Forward fill (column)

```python
data.fillna(method='ffill', axis='columns')
```

#### Backward fill (column)

```python
data.fillna(method='bfill', axis='columns')
```

#### Filling with Pandas objects
* There are many Pandas objects like df.sum(), df.max(), etc. we can fill the missing values withthese too.

```python
data.Temperature.fillna(data.Temperature.mean())
```

```python
data.Status.mode()
```

```python
data.Status.fillna('windy')
```

### Interpolate missing value
* In short, interpolation is a process of determining the unknown values that lie in between theknown data points. We can interpolate missing values based on different methods. This is doneby an object in DataFrame as interpolate() . By default, interpolate() does linear interpolation.

```python
data.Temperature
```

```python
data.Temperature.interpolate()
```

### Applying 'ffill' to a range of columns

```python
data.iloc[:, 0:3].fillna(method='ffill', axis='columns')
```
