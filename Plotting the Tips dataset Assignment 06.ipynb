### For Loop to plot Subplots

```python
for i in data_tips.columns:
    if data_tips[i].dtypes != 'object':
        sns.boxplot(y=data_tips[i])
        plt.show()
```

### Subplots in Matplotlib

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.boxplot(data=data_tips,y='tip')

plt.subplot(2,2,2)
sns.boxplot(data=data_tips,y='total_bill')

plt.subplot(2,2,3)
sns.boxplot(data=data_tips,y='sex')

plt.subplot(2,2,4)
sns.boxplot(data=data_tips,y='smoker')

plt.show()
```

### Correlation Matrix with Heatmap

```python
data_numerical = data_tips.select_dtypes(float)
data_numerical
```

```python
data_numerical.corr()
```

```python
sns.heatmap(data_numerical.corr(),annot=True)
plt.show()
```

```python
data_numerical.corr()[['tip']]
```

```python
sns.heatmap(data_numerical.corr()[["tip"]].sort_values(by="tip", ascending=False), vmin=-1, vmax=1,annot=True, cmap="Greens")
plt.plot();
```

```python
sns.heatmap(data_numerical.corr()[["total_bill"]].sort_values(by="total_bill", ascending=False), vmin=-1, vmax=1,annot=True, cmap="Reds")
plt.plot();
```
